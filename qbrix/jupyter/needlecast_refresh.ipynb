```python
import requests
import json
import subprocess
import os
from IPython.display import clear_output
```

```python
needlecast = "https://qbrix-core-internal.herokuapp.com/NeedleCast/authenticate"
```

```python
'''Enter the username'''
targetuser=input()
```

```python

url = needlecast

payload = json.dumps({
  "targetuserid": f"{targetuser}"
})
headers = {
  'Content-Type': 'application/json'
}

response = requests.request("POST", url, headers=headers, data=payload)
responseJson = json.loads(response.text)
token=responseJson['accessToken']
instanceurl=responseJson['instanceUrl']
userName=responseJson['details']['userName']


if userName is None or userName == '':
    userName = targetuser
```

```python
if not token is None and token != '':
    cmd = f"export SF_ACCESS_TOKEN='{token}' && sf org login access-token --instance-url  {instanceurl}  -a {userName} --no-prompt --json"
    subprocess.run([cmd], shell=True, capture_output=False)

clear_output(wait=False)

cmd = f"sf org list"
subprocess.run([cmd], shell=True, capture_output=False)
```
