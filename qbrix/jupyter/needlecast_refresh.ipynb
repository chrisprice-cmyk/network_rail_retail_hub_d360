```python
import requests
import json
import subprocess
import os 
```

```python
needlecast = "https://qbrix-core-internal.herokuapp.com/NeedleCast/authenticate"
if(os.environ.get("CODESPACES")=='true'):
    needlecast = "https://qbrix-runtime-service-8c3413c48d7f.herokuapp.com/NeedleCast/authenticate"
```

```python
'''Enter the username'''
targetuser=input()
```

```python

url = needlecast

payload = json.dumps({
  "targetuserid": f"{targetuser}"
})
headers = {
  'Content-Type': 'application/json'
}

response = requests.request("POST", url, headers=headers, data=payload)
responseJson = json.loads(response.text)
token=responseJson['accessToken']
instanceurl=responseJson['instanceUrl']
userName=responseJson['details']['userName']
                
```

```python
cmd = f"export SFDX_ACCESS_TOKEN='{token}' && sf org login access-token --instance-url  {instanceurl}  -a {userName} --no-prompt --json"
subprocess.run([cmd], shell=True, capture_output=False)
#cmd =f'sf org open -o {userName} --json'
#subprocess.run([cmd], shell=True, capture_output=False)
```
