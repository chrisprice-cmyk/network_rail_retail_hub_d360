```python
import requests
import json
import os
```

```python
gloopurl = "https://qbrix-core-stage-int.herokuapp.com/GLOOOP/orgtemplates"
if(os.environ.get("CODESPACES")=='true'):
    gloopurl = "https://qbrix-runtime-service-8c3413c48d7f.herokuapp.com/GLOOOP/orgtemplates"
```

```python
url = gloopurl
headers = {
  'Content-Type': 'application/json'
}
response = requests.request("GET", url, headers=headers)
print(response.text)
display(response.text)
templates =json.loads(response.text)

for template in templates:
    print(f'Name:{template["name"]}\n\tAlias:{template["alias"]}\n\tType:{template["type"]}\n\tAvailable:{template["availableCount"]}')

```
